```python
import pandas as pd
import numpy as np

chipo = pd.read_csv("chipolet.tsv", sep = '\t')
chipo

# 문자로된 item_price 컬럼을 전처리하여 float 형태로 변환하기
chipo["item_price_2"] = chipo["item_price"].apply(lambda x: float(x[1:]) )

chipo.head(10)
```

```python
# 각 컬럼 별로 그룹 연산을 다르게 적용하기.

adf = chipo.groupby(by="item_name").agg({
    "quantity": "sum",
    "item_price_2": "mean"
}).sort_values("quantity",ascending=False)

adf["profit"] = adf.apply(lambda x: x["quantity"] * x["item_price_2"],axis=1)
adf
```
